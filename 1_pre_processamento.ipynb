```python
import pandas as pd
import numpy as np
import re
```

```python
## dataset_url = https://archive.ics.uci.edu/ml/datasets/Diabetes+130-US+hospitals+for+years+1999-2008
data = pd.read_csv("diabetic_data.csv")
df = pd.DataFrame(data)
df
```

```python
df.columns
```

```python
for column in df.columns:
    index, counts = np.unique(df[column], return_counts=True)
    print(column, index, counts)
#     print(column, np.unique(df[column]))
    print("\n")
```

```python
df = df.drop(df[df["gender"] == "Unknown/Invalid"].index, axis=0)
```

```python
def formatAge(age):
    result = re.findall("\d+-\d+", age)[0].split("-")
    values = [int(n) for n in result]
    return int(np.mean(values))

df["age"] = df["age"].apply(lambda n: formatAge(n))
```

```python
def formatDiag(diag):
    if(diag == "?"):
        return diag
    elif(len(re.findall("[a-zA-Z]", diag)) == 0):
        return float(diag)
    else: 
        return "?"

df["diag_1"] = df["diag_1"].apply(lambda n: formatDiag(n))
df = df.drop(df[df["diag_1"] == "?"].index, axis=0)
```

```python
df["diag_2"] = df["diag_2"].apply(lambda n: formatDiag(n))
df = df.drop(df[df["diag_2"] == "?"].index, axis=0)
```

```python
df["diag_3"] = df["diag_3"].apply(lambda n: formatDiag(n))
df = df.drop(df[df["diag_3"] == "?"].index, axis=0)
```

```python
df.loc[df["max_glu_serum"] == "None", "max_glu_serum"] = -1
df.loc[df["max_glu_serum"] == "Norm", "max_glu_serum"] = 0
df.loc[df["max_glu_serum"] == ">200", "max_glu_serum"] = 1
df.loc[df["max_glu_serum"] == ">300", "max_glu_serum"] = 1
```

```python
df.loc[df["A1Cresult"] == "None", "A1Cresult"] = -1
df.loc[df["A1Cresult"] == "Norm", "A1Cresult"] = 0
df.loc[df["A1Cresult"] == ">7", "A1Cresult"] = 1
df.loc[df["A1Cresult"] == ">8", "A1Cresult"] = 1
```

```python
def defineLevel(col):
    df.loc[df[col] == "Down", col] = -1
    df.loc[df[col] == "No", col] = -1
    df.loc[df[col] == "Steady", col] = 0
    df.loc[df[col] == "Up", col] = 1
```

```python
define_level_cols = ["metformin", "repaglinide", "nateglinide", "chlorpropamide", "glimepiride", "acetohexamide",
                    "glipizide", "glyburide", "tolbutamide", "pioglitazone", "rosiglitazone", "rosiglitazone",
                    "acarbose", "miglitol", "troglitazone", "tolazamide", "insulin", "glyburide-metformin", 
                    "glipizide-metformin", "glimepiride-pioglitazone", "metformin-rosiglitazone", "metformin-pioglitazone"]

for col in define_level_cols:
    defineLevel(col)
```

```python
df.loc[df["change"] == "No", "change"] = 0
df.loc[df["change"] == "Ch", "change"] = 1
```

```python
df.loc[df["diabetesMed"] == "No", "diabetesMed"] = 0
df.loc[df["diabetesMed"] == "Yes", "diabetesMed"] = 1
```

```python
df.loc[df["readmitted"] == "NO", "readmitted"] = 0
df.loc[df["readmitted"] == "<30", "readmitted"] = 1
df.loc[df["readmitted"] == ">30", "readmitted"] = 1
```

```python
drop_columns = ["encounter_id", "patient_nbr", "race", "weight", "payer_code", "medical_specialty",
                "examide", "citoglipton"]

df = df.drop(drop_columns, axis=1)
```

```python
dummy_columns = ['gender', 'admission_type_id', 'discharge_disposition_id', 'admission_source_id']

df = pd.get_dummies(df, columns=dummy_columns)
```
